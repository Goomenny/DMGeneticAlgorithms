```python
import numpy as  np
from matplotlib import pyplot as plt
from sklearn.decomposition import PCA


x = [[0,-1],[1,0],[3,2],[4,3],[2,2],[2,0],[1.5,1.5],[1,1],[3.5,0.5],[2,1]]
x = np.array(x)


x = x-x.mean(axis=0)
```

```python
fig = plt.figure()
print(x)
L = np.array([1.0571,1])
plt.scatter(x[:,0],x[:,1])
z=L*x
plt.scatter(z[:,0],z[:,1])
plt.show()
```

```python
covmat = [[1.35,0.9],[0.9,1.25]]
covmat = np.array(covmat)
_, vecs = np.linalg.eig(covmat)

print(_,vecs)
fig = plt.figure()

origin = [0], [0] # origin point
myvec = np.array([[-0.6872,0.7264],[0.7264,0.6872]])

# plt.quiver(*origin, vecs[:,0], vecs[:,1],myvec, color=['r','b','g'], scale = 3)
plt.quiver(*origin, myvec[:,0], myvec[:,1], color=['r','b','g'], scale = 3)
plt.scatter(x[:,0],x[:,1])

newx= x.dot(myvec)
fig2 = plt.scatter(newx[:,0],newx[:,1])
plt.show()
```

```python
pca = PCA(n_components=1)

x=pca.fit_transform(x)
cov = pca.get_covariance()
print(cov)
fig = plt.plot(x)
plt.show()
print(x)
```
